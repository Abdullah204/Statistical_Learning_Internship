```python
import numpy as np                                    
from sklearn import datasets                                                        
from sklearn.linear_model import LinearRegression     
import matplotlib.pyplot as plt
from sklearn import preprocessing      
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
boston = datasets.load_boston()          
X = boston.data                          
y = boston.target  
M, N = X.shape
def adjR2 (M,N,R):
  return 1-(1-R)*(M-1)/(M-N-1)

#standardization :
# std_scaler = preprocessing.StandardScaler()
# X = std_scaler.fit_transform(X)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
idx = -1
score = 0
for i in range(N):
  reg = LinearRegression().fit(np.array([X.T[i]]).T, y)
  if(reg.score(np.array([X.T[i]]).T,y)>score):
    score = reg.score(np.array([X.T[i]]).T,y)
    idx = i
bm = np.array([X.T[idx]]).T
X = np.delete(X,idx,1)
#reg.score(bm,y)
#np.stack((a,X.T[0]),axis=1)
#np.delete(X,i,1)
#print(X.shape)
bestScore = score
print("best:",bestScore)
Z = X
while(True):
  M, N = Z.shape

  if(N == 0):
    break
  b = bm
  bestScore1 = 0
  delindex = -1
  for i in range(N):
    #print("i:",i)
    c = np.c_[b,Z.T[i]]
    reg = LinearRegression().fit(c, y)
    if(reg.score(c,y)>bestScore1):
      delindex = i
      bestScore1 = reg.score(c,y)
  #print("di:",delindex)
  b = np.c_[bm,Z.T[delindex]]
  Z = np.delete(Z,delindex,1)
  reg =LinearRegression().fit(b, y)
  m,n = b.shape
  print("score: ",adjR2(m,n,reg.score(b,y)),"best:" , bestScore)
  
  if adjR2(m,n,reg.score(b,y))<bestScore:
    break
  bestScore = adjR2(m,n,reg.score(b,y))
  bm = b



m,n = bm.shape
reg = LinearRegression().fit(bm, y)
print(adjR2(m,n,reg.score(bm,y)))
print(bm.shape)
print(reg.coef_)
```
